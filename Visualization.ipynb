```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
#line plot
```

```python
sns.lineplot(x='sepal_length',y='sepal_width',data=iris)
plt.title('line plot bw sepal length and width')
plt.show()
```

```python
sns.lineplot(x='sepal_length',y='sepal_width',data=iris,hue='species')
plt.title('line plot bw sepal length and width')
plt.show()
```

```python
sns.lineplot(x='tip',y='total_bill',data=tips)
plt.title('line plot bw tips and total bill')
plt.show()
```

```python
sns.lineplot(x='total_bill',y='tip',data=tips,hue='time')
plt.title('line plot bw tips and total bill')
plt.show()
```

```python
sns.lineplot(x='total_bill',y='tip',data=tips,hue='sex')
plt.title('line plot bw tips and total bill')
plt.show()
```

```python
sns.lineplot(x='total_bill',y='tip',data=tips,hue='day')
plt.title('line plot bw tips and total bill')
plt.show()
```

```python
sns.lineplot(x='total_bill',y='tip',data=tips,hue='smoker')
plt.title('line plot bw tips and total bill')
plt.show()
```

```python
#barplot
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('bar plot bw day and total bill')
plt.show()
```

```python
sns.barplot(x='species',y='sepal_length',data=iris)
plt.title('bar plot bw species and sepal length')
plt.grid(True)
plt.show()
```

```python
#barplot with different color
```

```python
sns.barplot(x='day',y='total_bill',data=tips,hue='sex',color='red')
plt.title('bar plot bw day and total bill')
plt.grid(True)
plt.show()
```

```python
#boxplot
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title('box plot bw day and total bill')
plt.show()
```

```python
#using hue
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
plt.title('box plot bw day and total bill')
plt.show()
```

```python
sns.boxplot(x='species',y='sepal_length',data=iris)
plt.title('box plot bw species and sepal length')
plt.show()
```
